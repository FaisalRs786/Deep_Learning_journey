```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv') # import data set
```

```python
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['iq'], hue=df['placement'])
```

```python
x = df.drop(columns=['placement','Unnamed: 0'])
y = df['placement']
```

```python
from sklearn.linear_model import Perceptron
pr = Perceptron()
pr.fit(x,y)
```

```python
pr.coef_
```

```python
pr.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf=pr)
```
