```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/DailyDelhiClimateTest.csv')
```

```python
df = df[['date','meantemp']]
df.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
plt.show()
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1.head()
```

```python
df['ewma'] = x1
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color = 'red')
plt.plot(df['date'],x1,color = 'blue',label = 'alpha = 0.9')
plt.legend()
plt.show()
```
